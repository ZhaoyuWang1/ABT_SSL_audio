```python
import json 
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
TASK_TYPES = [
    "environmental",
    "music",
    "speech",
    "other",
]
```

```python
with open("results.json", "r") as jsonfile:
    results = json.load(jsonfile)
```

```python
types = ["input_frames", "out_dim"]
data = {}
for type in types:
    data.setdefault(type, {})
    for k in results.keys():
        if type in k:
            value = int(k.split(":")[0].split("=")[-1][:-2])
            for task_type, scores in results[k].items():
                data[type].setdefault(task_type, {})
                data[type][task_type][value] = scores["AVERAGE"]
```

```python
for type, type_data in data.items():
    for task_type, scores in type_data.items():
        sorted_scores = sorted(scores.items())
        x = list(p for p, _ in sorted_scores)
        y = list(q for _, q in sorted_scores)
        plt.plot(x, y)
        plt.xlabel(f"{type}")
        plt.ylabel(f"average score ({task_type})")
        plt.tight_layout()
        plt.show()
```
